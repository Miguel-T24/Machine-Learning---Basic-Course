### Fase de prueba y de entrenamiento

```python
 from sklearn.model_selection import train_test_split
 from sklearn.linear_model import LinearRegression
```

```python
df_casas.head()
```

```python
df_casas.columns
```

```python
X = df_casas[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population']]
```

```python
X
```

```python
Y = df_casas['Price']
```

```python
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.3, random_state = 42)
```

```python
X_train
```

```python
 X_test
```

### Entrenando el modelo

```python
# Linear Regression Model
lrm = LinearRegression()
lrm.fit(X_train,Y_train)
```

### Evaluar el Modelo Entrenado

```python
# importamos los modulos
from sklearn import metrics
```

```python
# Prediccion
predictions = lrm.predict(X_test)
```

```python
# Es decir el valor de Y
predictions
```

```python
Y_test
```

```python
#Comparamos la prediccion con lo datos reales,mes decir, predictions con Y_test 
fig = px.scatter(x = Y_test,  y = predictions)
fig.show()
```

### El resultado del histograma debe ser una distribucion normal para ser aceptado el modelo

```python
fig = px.histogram(x = Y_test - predictions)
fig.show()  
```

Para evaluar de forma numerica el modelo utilizamos tres metricas
1. MAE (Mean Absolute Error) - Media del Valor Absoluto de los Errores
2.

MAE (Mean Absolute Error)- Media del valor Absoluto de los errores

**Mientras menor sea este valor, mejor es el modelo**

```python
metrics.mean_absolute_error(Y_test, predictions)
```

MSE (Media de los Errores al cuadrado)

```python
metrics.mean_squared_error(Y_test, predictions)
```

RMSE (Raiz cuadrada de la media d elos errores al cuadrado)

Es el mismo valor del MSE pero aplicandole la raiz cuadrada

```python
np.sqrt(metrics.mean_squared_error(Y_test, predictions))
```
